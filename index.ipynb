```python

import pandas as pd
import re

# Carrega o arquivo para Limpar e preparar os dados.

df = pd.read_csv('dataset/top_1000_imdb_movies.csv')

# Remover a coluna de índice extra
df = df.drop(columns=['Unnamed: 0'])

# Limpeza e Conversão da coluna 'Year of Release'
# Foi necessário extrair o ano (4 dígitos) para lidar com entradas como 'II 2018'
df['Year of Release'] = df['Year of Release'].astype(str).str.extract(r'(\d{4})', expand=False).astype(int)


df.head()
```

```python
# Criação de novo arquivo csv com os dados limpos
df.to_csv('dataset/cleaned_top_1000_imdb_movies.csv', index=False)

```
